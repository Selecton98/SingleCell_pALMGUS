```r
df<-read.table('GSE175384_Counts.txt',sep='\t')
```

```r
colnames(df)<-df[1,]
df<-df[-1,]
rownames(df)<-df[,1]
df<-df[,-1]
```

```r
colnames(df)
```

```r
df.1<-df[,paste('RNAseq',c(1:32,74:79),sep='')]
```

```r
df.1
```

```r
for (i in 1:38){
df.1[,i]<-as.numeric(as.character(df.1[,i]))}
```

```r
name<-read.table('proteincodinggenename.txt',sep='\t')
```

```r
df.1<-df.1[name$x,]
```

```r
df.1<-df.1[-grep('MT-', row.names(df.1)),]
df.1<-df.1[-grep('RPL', row.names(df.1)),]
df.1<-df.1[-grep('RPS', row.names(df.1)),]
```

```r
type<-c(rep('AL',times=32),rep('MGUS',times=6))
```

```r
require(dplyr)
require(tidyr)
library(DESeq2)
```

```r
# 这一步很关键，要明白condition这里是因子，不是样本名称；小鼠数据有对照组和处理组，各两个重复

condition <- factor(type, levels = c("AL","MGUS"))
condition
#colData也可以自己在excel做好另存为.csv格式，再导入即可
colData <- data.frame(row.names=colnames(df.1), condition)
colData

##############################

dds2 <- DESeqDataSetFromMatrix(df.1, colData, design=~condition)
dds2 <- DESeq(dds2)

res= results(dds2)
res = res[order(res$pvalue),]
head(res)
row.names(res)
summary(res)
#所有结果先进行输出
table(res$padj<0.05)
```

```r
normalized_counts <- counts(dds2, normalized=TRUE)
```

```r
dim(normalized_counts)
```

```r
options(repr.plot.height=6,repr.plot.width=6)
```

```r
library(ggplot2)
library(ggfortify)
library(devtools)
know2<-as.data.frame(t(normalized_counts))
know2$type<-type
autoplot(prcomp(know2[,1:17897]), data=know2,label=TRUE,label.size=4,colour="type")+theme_bw()
```
