```r
allPC4<-subset(allPC3,cells=anno3[,1])
```

```r
allPC4
```

```r
matrix <- allPC4@assays$RNA@counts
```

```r
infercnv_obj = CreateInfercnvObject(raw_counts_matrix=matrix, # 可以直接提供矩阵对象
                                    annotations_file="0925-integrated.anno-2.txt",
                                    delim="\t",
                                    gene_order_file="genecode_v22_pos.sorted.inuse.bed",
                                    ref_group_names=c('normalPC_0','normalPC_1','normalPC_2','normalPC_3','normalPC_4','normalPC_5','normalPC_6','publicPC_0','publicPC_1','publicPC_2','publicPC_3','publicPC_4','publicPC_5','publicPC_6'))
```

```r
infercnv_obj = infercnv::run(infercnv_obj,
                             cutoff=0.1,  # use 1 for smart-seq, 0.1 for 10x-genomics
                             out_dir="0925output-2_dir",  # 输出文件夹
                             cluster_by_groups=T,   # 聚类
                             denoise=T, #去噪
                             HMM=T) # 是否基于HMM预测CNV
```
